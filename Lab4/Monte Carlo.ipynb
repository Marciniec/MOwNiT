```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(np.arange(0.00001, 1.0, 0.000001))
plt.plot(x,x)
plt.show()
```

```python
calls = []
errors = []
with open('data/monte_plain.csv') as inputFile:
    for line in inputFile:
        call, error = line.split(", ")
        calls.append(call)
        errors.append(error)
```

```python
calls =list(map(lambda x: float(x),calls))
errors =list(map(lambda x: float(x[:-2]),errors))
```

```python
plt.figure(figsize=(18,9))
plt.plot(calls, errors,'o')
plt.title('Monte Carlo PLAIN')
plt.xlabel('calls')
plt.ylabel('error')
plt.show()
```

```python
divisions = []
errors = []
with open('data/przedzialy.csv') as inputFile:
    for line in inputFile:
        error, division = line.split(",")
        divisions.append(division)
        errors.append(error)
```

```python
divisions =list(map(lambda x: float(x),divisions))
errors =list(map(lambda x: float(x[:-2]),errors))
```

```python
plt.figure(figsize=(18,9))
plt.plot(divisions, errors,'o')
plt.xlabel('divisions')
plt.ylabel('error')
plt.xscale('log')
# plt.yscale('linear')
plt.show()
```
